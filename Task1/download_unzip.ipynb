```python
for filename in os.listdir('A:\\Users\\Sam\\Downloads\\'):
    if '.zip' in filename and '答卷汇总' in filename:
        for aid, atitle in aid2atitle.items():
            if atitle in filename:
                shutil.move('A:\\Users\\Sam\\Downloads\\%s' % filename, 'zips/%s.zip' % aid)
                break
```

```python
if not os.path.exists('raws'):
    os.makedirs('raws')
for filename in os.listdir('zips'):
    base, ext = os.path.splitext(filename)
    if ext == '.zip':
        subprocess.call(['unzip', '-o', 'zips/%s' % filename])
        for filename2 in os.listdir('.'):
            if os.path.splitext(filename2)[1] == '.csv':
                shutil.move('./%s' % filename2, 'raws/%s.csv' % base)
```
